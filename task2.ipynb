```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Cognizfy\Task 2\Dataset .csv")
df.head()
```

```python
df = df[['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Price range']]
```

```python
df = df.dropna()
```

```python
df_encoded = pd.get_dummies(df[['Cuisines', 'Average Cost for two', 'Price range']], drop_first=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(df_encoded)
```

```python
def recommend(name):
    idx = df[df['Restaurant Name'] == name].index[0]

    distances = list(enumerate(similarity[idx]))

    distances = sorted(distances, key=lambda x: x[1], reverse=True)

    print("Recommended Restaurants:")
    for i in distances[1:6]:
        print(df.iloc[i[0]]['Restaurant Name'])
```

```python
recommend("Cafe Coffee Day")
```

## Restaurant Recommendation System

### Objective:
To recommend restaurants based on similarity in features like cuisines, price range, and cost.

### Data Preprocessing:
- Selected relevant features: Restaurant Name, Cuisines, Cost, Price range
- Removed missing values
- Converted categorical data into numerical format using one-hot encoding

### Method Used:
- Cosine Similarity was used to measure similarity between restaurants

### How it Works:
- Each restaurant is compared with all others
- Similarity scores are calculated
- Top 5 most similar restaurants are recommended

### Conclusion:
The system successfully recommends restaurants based on user preference and similarity in features.

This approach is a content-based filtering method, commonly used in real-world recommendation systems.
